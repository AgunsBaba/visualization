This project is to visualize information about the 20 biggest solar power plants in the world

```python
#import modules
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
#import data
df = pd.read_excel('/content/20 Largest Solar Power Plants.xlsx')
df.head()
```

```python
#rename columns
df.rename(columns = {'MWp (Megawatts Peak Capacity)': 'mwp', 'Sq Mi': 'sqmi', 'Sq Km': 'sqkm'}, inplace=True)
df.head()
```

```python
df.info()
```

```python
#check the number of plants in each country
df.Country.value_counts()
```

```python
#get total power generated by each country
power_by_country = df.groupby(['Country'])['mwp'].sum()
power_by_country
```

```python
#put the countries and the power generated by each into a list
country_list = list(power_by_country.index)
country_power = list(power_by_country.values)
```

```python
#create a basic bar chart to view the total power generated by country
plt.style.use('seaborn')
plt.bar(country_list, country_power)
```

```python
#rotate the labels so that they're readable
plt.bar(country_list, country_power)
plt.xticks(rotation=45, ha='right')
```

```python
#use different colors for the bars
plt.bar(country_list, country_power, color=['black', 'red', 'blue', 'cyan', 'magenta', 'green'])
plt.xticks(rotation=45, ha='right')
```

```python
#set title for the chart
plt.bar(country_list, country_power, color=['black', 'red', 'blue', 'cyan', 'magenta', 'green'])
plt.xticks(rotation=45, ha='right')
plt.xlabel('Countries', fontsize=16)
plt.ylabel('Power Generated', fontsize=16)
plt.title('Total Power Generated by Country', fontsize=16)
```
